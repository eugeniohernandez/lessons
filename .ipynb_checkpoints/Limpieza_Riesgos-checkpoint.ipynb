```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('Riesgos.xlsx')
```

```python
data.head()
```

```python
dic = {'EL': [],'OT':[],'BU':[],'FS':[],'HR':[],'FA':[],'IT':[],'PT':[],
       'TR':[]}

for i in data['Risk']:
    a = str(i).split(';')
    for j in a:
        if j[0:2] in dic:
            dic[j[0:2]].append(j)

max=533
for i in dic:
    if len(dic[i])<max:
        for j in range(max-len(dic[i])):
            dic[i].append(' ')
```

```python
data_final = pd.DataFrame(dic)
data_final.head()
```

```python
data_final.to_excel('si.xlsx')
```
