```python
data=pd.read_csv('./files_for_lab/marketing_customer_analysis.csv')    # import csv file
```

```python
data.head(1)
```

```python
def clean_data(df):
    df.columns=[e.lower().replace(' ', '_') for e in df.columns]
    df=df.drop(columns=['vehicle_size'])
    df=df.dropna()
    df['effective_to_date']=pd.to_datetime(df['effective_to_date'], errors='coerce')
    return df
```

```python
clean_data(data) # call the function
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('./files_for_lab/marketing_customer_analysis.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
# Total number of respones

```

```python
# Response rate by sales channel

plt.figure(figsize=(8,4))
sns.countplot('Sales Channel', hue='Response', data=data)
plt.ylabel('Response by Sales Channel')
plt.show();
```

```python
#Show a plot of the response rate by the total claim amount

plt.figure(figsize=(12,6))
sns.boxplot(y='Total Claim Amount' , x='Response', data=data)
plt.ylabel('Response by Total Claim Amount')
plt.show();
```

```python
#Show a plot of the response rate by income

plt.figure(figsize=(12,6))
sns.boxplot(y='Income' , x='Response', data=data)
plt.ylabel('Response by Inncome')
plt.show();
```
