## Preguntas

### Agregaciones simples

#### Qué procentaje de usuarias registradas han comprado al menos un producto?

```python
%%sql buyer_transactions <<
select buyer_id, count(distinct transaction_id) as transactions, register date
from users u
left join transactions t
on u.user_id = t.buyer_id
group by user_id
```

```python
buyer_transactions['transactions'] = list(map(lambda x: 1 if x > 0 else 0, buyer_transactions['transactions']))
round(buyer_transactions['transactions'].mean()*100,2)
```

#### Qué procentaje de compradoras no están registradas?

```python
%%sql all_transactions <<
select *
from transactions t
left join users u
on t.buyer_id = u.user_id
group by transaction_id
```

```python
all_transactions.register_date.count(None)/len(all_transactions)*100
```

### Retención externa

#### Qué procentaje de usuarias han comprado al menos un producto el mes en el que se registró?

```python
%%sql retention <<
select 
        user_id,
        strftime('%m', register_date) as month,
        strftime('%Y-%m-01',register_date) as register_month, 
        strftime('%Y-%m-01', transaction_date) as transaction_month,
        round((julianday(strftime('%Y-%m-01', transaction_date)) - julianday(strftime('%Y-%m-01',register_date))) /30)  as datediff
        
from users u
left join transactions t
on u.user_id = t.buyer_id
```

```python
retention
```

```python
retention.groupby(['month','datediff'])['transaction_month'].count().unstack()
```
